## 2.2 Dataset Information

- **userWallet**: Wallet address of the user -> (e.g., 0x000000000096026fb41fc39f9875d164bd82e2dc)

- **network**: Blockchain network used for the transaction -> (e.g., polygon, ethereum)

- **protocol**: DeFi protocol involved -> (e.g., aave_v2)

- **txHash**: Unique transaction hash on the blockchain -> (e.g., 0xe2d7eb815c89331a734ed6f204a06c385a1b39040baadf59c3b29099fb138132)

- **logId**: Unique identifier combining transaction hash and action 

- **timestamp**: Time of the transaction 

- **blockNumber**: Block number of the transaction -> (e.g., 1627118913)

- **action**: Type of interaction with the protocol -> (e.g., deposit, borrow, repay, redeemunderlying, liquidationcall)

- **actionData**: Detailed information about the action 

  - **type**: Specific action type -> (e.g., Deposit, Borrow)
  - **amount**: Amount of asset transacted -> (e.g., 1000000000000000 in wei)
  - **assetSymbol**: Symbol of the asset -> (e.g., WMATIC)
  - **assetPriceUSD**: Price of the asset in USD -> (e.g., 0.9223772540040778087468127454060907)
  - **poolId**: Contract address of the liquidity pool -> (e.g., 0x0d500b1d8e8ef31e21c99d1db9a6444d3adf1270)
  - **userId**: Wallet address linked to the action -> (e.g., 0x000000000096026fb41fc39f9875d164bd82e2dc)

- **createdAt**: Time the record was created in the database 

- **updatedAt**: Time the record was last updated in the database 

- **creditScore**: Predicted credit score based on transaction behavior 


## Rationale for Dropping Columns

- **network**: Dropped as its impact on wallet behavior is minimal and not critical for credit scoring.

- **protocol**: Dropped since it contains only 'aave_v2', offering no variability for analysis.

- **txHash**: Dropped as it’s a unique identifier, irrelevant to behavioral patterns.

- **logId**: Dropped as it’s a derived identifier, redundant with `txHash` and `action`.

- **poolId**: Dropped as it’s contract-specific and unrelated to creditworthiness.

- **userId**: Dropped as it duplicates `userWallet`, providing no additional value.

- **createdAt**: Dropped as it’s database metadata, not reflective of transaction behavior.

- **updatedAt**: Dropped as it’s database metadata, irrelevant to scoring analysis.

```python


plt.figure(figsize=(10, 5))
sns.countplot(x='action', data=df, order=df['action'].value_counts().index)
plt.title('Distribution of Actions')
plt.ylabel('Count')
plt.xlabel('Action Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


plt.figure(figsize=(10, 5))
sns.boxplot(x='action', y='amount', data=df)
plt.yscale('log')  # log scale because DeFi values vary a lot
plt.title('Amount Distribution by Action Type')
plt.xlabel('Action')
plt.ylabel('Amount (Log Scale)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['date'] = df['timestamp'].dt.date

action_over_time = df.groupby(['date', 'action']).size().reset_index(name='count')

plt.figure(figsize=(14, 6))
sns.lineplot(data=action_over_time, x='date', y='count', hue='action')
plt.title('Transaction Trends Over Time by Action')
plt.ylabel('Daily Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_assets = df.groupby('assetSymbol')['amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_assets.index, y=top_assets.values)
plt.title("Top 10 Assets by Total Amount Interacted With")
plt.ylabel("Total Amount (raw units)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='assetPriceUSD', y='amount', alpha=0.3)
plt.title("Transaction Amount vs Asset Price")
plt.xlabel("Asset Price (USD)")
plt.ylabel("Amount")
plt.tight_layout()
plt.show()
```

```python
txn_counts = df.groupby('userWallet').size()

plt.figure(figsize=(10, 6))
sns.histplot(txn_counts, bins=50, log_scale=True)
plt.title("Distribution of Transaction Counts per Wallet (Log Scale)")
plt.xlabel("Number of Transactions")
plt.ylabel("Number of Wallets")
plt.tight_layout()
plt.show()
```

```python
action_volume = df.groupby('action')['amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=action_volume.index, y=action_volume.values)
plt.title("Total Amount Moved per Action Type")
plt.ylabel("Cumulative Amount")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Univariate analysis : Distribution of amount

```python
sns.set(style='whitegrid')


f, ax = plt.subplots(1, 2, figsize=(20, 10))

# Countplot (bar)
sns.countplot(x=df['action'], data=df, palette='bright', ax=ax[0], saturation=0.95)
ax[0].set_title("Action Count", fontsize=18)
ax[0].set_xlabel("Action Type", fontsize=14)
ax[0].set_ylabel("Count", fontsize=14)
ax[0].tick_params(axis='x', labelrotation=30)


for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=14)

# Pie chart
explode = [0.1 if i == 0 else 0 for i in range(df['action'].nunique())] 
ax[1].pie(
    x=df['action'].value_counts(),
    labels=df['action'].value_counts().index,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    startangle=140,
    colors=sns.color_palette('bright')
)
ax[1].set_title("Action Distribution (Pie Chart)", fontsize=18)

plt.tight_layout()
plt.show()
```

## Bivariate Analysis: action vs amount

```python
grouped_data = df.groupby('action')

f, ax = plt.subplots(1, 3, figsize=(24, 8))

sns.barplot(
    x=grouped_data['amount'].mean().index,
    y=grouped_data['amount'].mean().values,
    palette='mako',
    ax=ax[0]
)
ax[0].set_title('Average Amount by Action', color='#005ce6', size=20)
ax[0].tick_params(axis='x', labelrotation=30)
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=13)

sns.barplot(
    x=grouped_data['assetPriceUSD'].mean().index,
    y=grouped_data['assetPriceUSD'].mean().values,
    palette='flare',
    ax=ax[1]
)
ax[1].set_title('Average Asset Price (USD) by Action', color='#005ce6', size=20)
ax[1].tick_params(axis='x', labelrotation=30)
for container in ax[1].containers:
    ax[1].bar_label(container, color='black', size=13)


sns.barplot(
    x=grouped_data['blockNumber'].mean().index,
    y=grouped_data['blockNumber'].mean().values,
    palette='coolwarm',
    ax=ax[2]
)
ax[2].set_title('Average Block Number by Action', color='#005ce6', size=20)
ax[2].tick_params(axis='x', labelrotation=30)
for container in ax[2].containers:
    ax[2].bar_label(container, color='black', size=13)

plt.tight_layout()
plt.show()
```

## Does Action Type Affect Amount?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='action', y='amount', palette='Set2')
plt.title("Does Action Type Impact Amount Transacted?")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

## Are Liquidated Wallets Riskier?

```python

liquidated_wallets = df[df['action'] == 'liquidationcall']['userWallet'].unique()
df['is_liquidated'] = df['userWallet'].isin(liquidated_wallets)

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='is_liquidated', y='amount', palette='coolwarm')
plt.title("Are Liquidated Wallets Doing Higher Amount Transactions?")
plt.xticks([0, 1], ['No', 'Yes'])
plt.tight_layout()
plt.show()
```

## Does Wallet Activity Level Impact Amounts?

```python
wallet_stats = df.groupby('userWallet').agg({
    'amount': ['mean', 'count']
}).reset_index()
wallet_stats.columns = ['userWallet', 'avg_amount', 'txn_count']

plt.figure(figsize=(10, 6))
sns.scatterplot(data=wallet_stats, x='txn_count', y='avg_amount', alpha=0.5)
plt.xscale('log')  # helps with skew
plt.title("Do High-Activity Wallets Transact Higher Amounts?")
plt.tight_layout()
plt.show()
```
